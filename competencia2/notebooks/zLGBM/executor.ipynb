```r
# Convert notebook to R script and source it
execute_notebook_r <- function(notebook_path) {
  # Convert ipynb to R
  system(paste("jupyter nbconvert --to script", notebook_path))
  
  # Get the R script path
  r_script <- sub("\\.ipynb$", ".R", notebook_path)
  
  # Source the R script
  if (file.exists(r_script)) {
    source(r_script)
    cat("Successfully executed", notebook_path, "as R script\n")
  } else {
    cat("Failed to convert notebook to R script\n")
  }
}

execute_notebook_r("/apo-501.ipynb")
```

```r
# Find where jupyter is installed
system("which jupyter")  # On Linux/Mac

# Then use the full path
system("/usr/local/bin/jupyter nbconvert --execute --to notebook --inplace apo-501.ipynb")
```
