# Desde acá puedo llamar a todos las notebooks

```r
library(reticulate)

execute_notebook_direct <- function(notebook_path) {
  # Use nbconvert Python API directly - no shell commands needed
  nbconvert <- import("nbconvert")
  json <- import("json")
  
  tryCatch({
    # Create an executor
    executor <- nbconvert$execute$executor$NotebookExecutor()
    
    # Read the notebook
    with(import_builtins()$open(notebook_path, 'r'), function(f) {
      notebook_content <- json$load(f)
    })
    
    # Execute the notebook
    executor$execute(notebook_content, resources = list())
    
    # Write the executed notebook back
    with(import_builtins()$open(notebook_path, 'w'), function(f) {
      json$dump(notebook_content, f)
    })
    
    cat("✅ Successfully executed", notebook_path, "using direct Python API\n")
    
  }, error = function(e) {
    cat("❌ Error:", e$message, "\n")
  })
}

# Use it
execute_notebook_direct("apo-501.ipynb")
```
